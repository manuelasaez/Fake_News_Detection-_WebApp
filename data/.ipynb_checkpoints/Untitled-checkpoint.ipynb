```python
import pandas as pd


train_df = pd.read_csv('train.csv', dtype='str')
train_df.dtypes

# Convert non-string objects in strings for title, author and text
train_df['title'] = train_df['title'].astype(str)
train_df['author'] = train_df['author'].astype(str)
train_df['text'] = train_df['text'].astype(str)

# Lowercase text data for title, author and text columns
train_df['title'] = train_df['title'].str.lower()
train_df['author'] = train_df['author'].str.lower()
train_df['text'] = train_df['text'].str.lower()

# Create new column called "text_title" merging title and text columns
train_df['text_title'] = train_df['text']+train_df['title']


```
